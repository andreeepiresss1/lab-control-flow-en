## Words

```python
words = ['play', 'filling', 'bar', 'theatre', 'easygoing', 'date', 'lead', 'that', 'story',  'island']
```

**Print every word in upper case**

```python
# your code here
for each_word in words:
    print(each_word.capitalize())
```

**Create a new list containing only words with 5 or more letters**

```python
# your code here
new_list1 = []
for each_word in words:
    if len(each_word) >= 5:
        new_list1.append(each_word)

print(new_list1)
```

**Print the first word starting with "t"**

```python
# your code here
for each_word in words:
    if each_word[0] == "t":
        print(each_word)
```

## Numbers

**Create a list containing the square of every number from 1 to 10**

```python
# your code here
new_list2 = []
for each_number in range (11):
    new_list2.append(each_number**2)

print(new_list2)
```

**Print a list containing the square of every odd number from 1 to 10**

```python
# your code here
new_list3 = []
for each_number in range (11):
    if each_number % 2 != 0:
        new_list3.append(each_number**2)

print(new_list3)
```

**Create a list with the squares of all multiples of 8 below 1000**

## People

```python
people = [
    {
        "name": "Juan",
        "age": 34,
        "n_kids": 2
    },
    {
        "name": "Pepe",
        "age": 27,
        "n_kids": 0
    },
    {
        "name": "Sonia",
        "age": 41,
        "n_kids": 1
    },
    {
        "name": "Lucía",
        "age": 22,
        "n_kids": 2
    },
    {
        "name": "Leo",
        "age": 55,
        "n_kids": 5
    }
]
```

**How many people are there?**

```python
# your code here
len(people)
```

**How many people have kids**?

```python
# your code here
nb_people_with_kids = 0
for each_person in people:
    if each_person['n_kids'] > 0:
        nb_people_with_kids += 1

print(nb_people_with_kids)
```

**How many kids do they have in total?**

```python
# your code here
nb_kids = 0
for each_person in people:
    nb_kids += each_person['n_kids']

print(nb_kids)
```
